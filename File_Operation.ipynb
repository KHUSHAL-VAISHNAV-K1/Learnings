File Operation 

```python
# Reading a whole file 
with open('tx2.txt','r') as file:
    t=file.read()
    print(t)
```

```python
# Read a file line by line 
with open('tx2.txt', 'r') as file:
    for line in file:
        print(line.strip())

# without .strip() - show the new line at each stage, by leaving the line every time 
# With .strip() - show the format of text same as in the file 
```

```python
# open a file to write inside it. 
with open('text.txt','w') as file:
    file.write('Hello World\n') 

# Entire content is overwritten.
```

```python
# Writing a file without overwriting. 
with open ('text.txt','a') as file:
    file.write("Append Approach\n")
    
```

```python
# Writing a list of lines to a file 
lines = ["First line \n", "Second line \n", "Third line\n"]
with open('text.txt','a') as file:
    file.writelines(lines)
```

```python
# Working with Binary File 
data = b'\x00\x01\x02\x03'
with open('binary.bin','wb')as file:
    file.write(data)
```

```python
# Reading a binary file 
data = b'\x00\x01\x02\x03'
with open('binary.bin','rb')as file:
    bit = file.read()
    print(bit)
```

```python
# Read the content form a source text file and write to a destination text file 
# copying a text file 
with open('tx2.txt','r') as source_file:
    content = source_file.read()

with open('new file.text','w') as destination_file:
    destination_file.write(content)
```

```python
# Read a text file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file
with open('tx2.txt','r') as file:
    lines = file.readlines()
    line_count = len(lines)
    word_count = sum(len(line.split()) for line in lines)
    char_count = sum(len(line) for line in lines)

print(f"line_count : {line_count}  \nword_count : {word_count} \nchar_count : {char_count}")
```
