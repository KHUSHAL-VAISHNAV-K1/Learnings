```python
# Data Serialisation - 
import json
data = {"name" : "khushal", "age" : 23}
converted = json.dumps(data)
print(converted)
# returning json in string format 
print(type(converted))

last_type = json.loads(converted)
print(type(last_type))
print(last_type)
```
