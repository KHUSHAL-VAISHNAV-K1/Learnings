```python
# Creating a new class
class Car:

    def __init__(self,Brand,Price):
        self.Brand = Brand
        self.Price = Price

    def dis(self):
        print("Brand : ", self.Brand)
        print("Price : ", self.Price)
car1 = Car("Audi" , 90000)
car2 = Car("Bmw" , 845189)    

car1.dis()
car2.dis()
```
