```python
# Printing a complete dictionary
Student = {"name ": "krish","age ": 21,"grade":23}
print(Student)
```

```python
# Error case - key with same name will be overwritten.
Stud = {"name ": "krish","age ": 21,"name ":23}
print(Stud)
```

```python
# Accessing dictionary element 
dic = {"name ": "krish","age ": 21,"grade":"a"}
print(dic['grade'] )

# Other method - get()
print(dic.get('age '))

# If the key is doesn't exist this will print the message.
print(dic.get('last_element',"Doesn't exist")) 

```

```python
dic = {"name ": "krish","age ": 21,"grade":"a"}
# Modifying dictionary element

# Updating key
dic["age "]=33
print(dic)

# Adding a new key and value
dic["Address"]= "jaipur"
print(dic)
```

```python
# Deleting a key 
del dic["Address"]
print(dic)
```

```python
# Dictionary Methods:

# getting all the keys and values indivisually
key=dic.keys()
print(key)
values=dic.values()
print(values)

# Getting it combined
item = dic.items()
print(item)
```

```python
# Shallow copy - use ".copy() this will copy the dictionary and any updates in previous dictionary will not take place in the new one."

dic = { "name": "krish" , "age" : 21, "grade" : "a" }

dic_copy=dic
print(dic)

print(dic_copy)
dic["name"] = "khushal "


dic1=dic.copy()

print(dic1)
```

```python
# Iteration over dictionaries

for i in dic.keys():
    print(i)


for k in dic.values():
    print(k)




for keys, v in dic.items():
    print(keys, v)

```

```python
# Nested dictionaries
mca = {
    "s1": {"name": "A", "age": 20},
    "s2": {"name": "B", "age": 23}
}

print(mca["s1"]["name"])

# Iteration over nested dictionaries
for name in mca.items():
    print(name)
```

```python
# Nested dictionaries -        important hai
mca = {
    "s1": {"name": "A", "age": 20},
    "s2": {"name": "B", "age": 23}
}

print(mca["s1"]["name"])

# Iteration over nested dictionaries
for k,v in mca.items():
    print(f"{k}:{v}")
    for key,value in v.items():
        print(f"{key}:{value}")
```

```python
# Dictionary Comprehension 
comp = {i: i for i in range(50) if i%5==0}
print(comp)
```
