Inheritance In OOPS

```python
# Creating parent class to understand the inheritance. 
class car : 
    def __init__(self, Brand,Price):
        self.Brand = Brand
        self.Price = Price

    def drive(self):
        print(f"He will drive {self.Brand}")
```

```python
car1 = car("Audi",9999)
car1.drive()

# "None" will return in case of funtion when this is not return anything.
```

```python
# Inherit property
class new_car(car):
    def __init__(self,Brand,Price,is_selfdriving):
        super().__init__(Brand,Price)
        self.is_selfdriving = is_selfdriving
    def selfdrive(self):
        print(f"new car is Tesla supports self driving {self.is_selfdriving}")
```

```python
Tesla = new_car("Tesla",100,True)
Tesla.selfdrive()
```

```python
car1.drive()
Tesla.drive()
```

```python
# Multiple inheritance 
# When a class inherits more than one base class 
 
class animal :
    def __init__(self,name):
        self.name = name 

    def speak(self):
        print("Subclasses must implement this method")

# Base class 2 
class Pet:
    def __init__(self,owner):
        self.owner = owner

# Derived Class 
class Dog(animal,Pet):
    def __init__(self, name,owner):
        animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} say wolf"
    
# Creating an object 
Wolf = Dog("Mickey","Khush")
Wolf.speak()
```
